```python
df = df.drop(['year'],axis=1)
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,12]
y
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
Fit_feature = SelectKBest(score_func = f_classif)
```

```python
Fit_feature.fit(x,y)
```

```python
pd.DataFrame(Fit_feature.scores_)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state= 1)
xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rfl = RandomForestClassifier()
```

```python
Rfl = Rfl.fit(xtrain,ytrain)
Rfl
```

```python
Rfl.score(xtest,ytest)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

# creatin decision tree obj


```python
Dcf = DecisionTreeClassifier()
```

```python
Dcf.fit(xtrain,ytrain)
```

```python
Dcf.score(xtest,ytest)
```
